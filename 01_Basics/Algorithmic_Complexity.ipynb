# Counting Operation.

--Assume these steps take

  Constanst time:


*   Mathemetical opertions
*   comparsions.
*   assignments.
*   accessing object in memory
*   then count the number of operations executed as funtion of size of input.



```python
def c_to_f(c):
  return c*9.0/5 +32

def mysum(x):
  total = 0

  for i in range(x+1):
    total += i
  return total
```

```python
c_to_f(12)
```

```python
mysum(10)
```
