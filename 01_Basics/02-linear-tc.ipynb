# Learn with time complexity O(n)

```python
def inToStr(i):
    digits= '0123456789'
    if i == 0:
        return '0'

    result= ''
    while i > 0:
        result = digits[i%10] + result
        i = i//10

    return result

print(inToStr(123))
```

**time complexity o(n square 2) or space complexity O(n)**

```python
n = 1000
k = 0

for i in range(n // 2, n + 1):  # i = n/2 to n
    j = 2
    while j < n:  # j = 2; j < n; j *= 2
        k += n // 2
        j *= 2

print(k)
```

# Total complexity

Outer × Inner → O(n/2 * log n) → O(n log n)

```python
def binary_search_recursive(arr, target, left, right):
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search_recursive(arr, target, mid + 1, right)
    else:
        return binary_search_recursive(arr, target, left, mid - 1)
```

# Time Complexity

Same as iterative → O(log n)

# Space Complexity

Recursion adds log n stack calls → O(log n)

```python
L = [1, 2, 3, 4, 5]

for i in range(0, len(L)):
    for j in range(i+1, len(L)):
        print('({}, {})'.format(L[i], L[j]))
```
